```python
import pandas as pd
```

```python
books = pd.read_csv('../data/cleaned_amzn/books_final.csv')
books.head()
```

```python
def capitalize_each_word(s):
    try :
        return s.title()
    except:
        return s
```

```python
def convert_title(s):
    # Define a set of words to exclude from capitalization
    exclusions = {"a", "and", "the", "of", "in", "on", "for", "to", "but", "with"}
    
    try:
        # Split the string into words
        words = s.split()
        
        # Capitalize each word according to the rules
        title_cased = []
        for i, word in enumerate(words):
            if word.lower() not in exclusions or i == 0 or i == len(words) - 1:
                title_cased.append(word.capitalize())
            else:
                title_cased.append(word.lower())
        
        # Join the words back into a single string
        return ' '.join(title_cased)
    except:
        return s
```

```python
books['Title'] = books['Title'].apply(convert_title)
books.head()
```

```python
books["authors"] = books["authors"].apply(capitalize_each_word)
books.head()
```

```python
def sentence_case(s):
    try:
        output = ""
        isFirstWord = True

        for c in s:
            if isFirstWord and not c.isspace():
                c = c.upper()
                isFirstWord = False
            elif not isFirstWord and c in ".!?":
                isFirstWord = True
            else:
                c = c.lower()

            output = output + c

        return output
    except:
            return s

books["description"] = books["description"].apply(sentence_case)
books.head()
```

# Ratings

```python
ratings = pd.read_csv('../data/cleaned_amzn/ratings_final.csv')
ratings.head()
```

```python
ratings_concise = ratings[['User_id', 'book_id', 'review/score']]
ratings_concise = ratings_concise.rename(columns={'review/score': 'rating'})
# ratings_concise = ratings_concise.sort_values(by=['User_id'])
ratings_concise.head()
```

```python
# Reindex starting from 1
ratings_concise.index = range(1, len(ratings_concise) + 1)
```

```python
ratings_concise.head()
```

```python
# ratings_concise.to_csv('../data/cleaned_amzn/ratings_final_final.csv', index=False)
```

```python
# Create new user ids
user_ids = ratings_concise['User_id'].unique()
user_id_map = {user_id: i for i, user_id in enumerate(user_ids)}
ratings_concise['new_user_id'] = ratings_concise['User_id'].apply(lambda x: user_id_map[x])
ratings_concise.head()
```

```python
ratings_concise.drop(columns=['User_id'], inplace=True)
ratings_concise = ratings_concise.rename(columns={'new_user_id': 'User_id'})
ratings_concise.head()
```
