```python
# import fuzz
from fuzzywuzzy import fuzz


# Function to display similar titles within the same author group
def display_similar_titles(df, threshold=80):
    similar_titles = []

    grouped = df.groupby('authors')
    for author, group in grouped:
        titles = group['Title'].tolist()
        # print(titles)
        title_ids = group['book_id'].tolist()

        for i in range(len(titles)):
            for j in range(i + 1, len(titles)):
                title1 = titles[i]
                title2 = titles[j]
                similarity = fuzz.ratio(title1, title2)

                if similarity > threshold:
                    similar_titles.append((title_ids[i], title1, title_ids[j], title2, similarity))

    # Display similar titles
    similar_titles_df = pd.DataFrame(similar_titles, columns=['ID1', 'Title1', 'ID2', 'Title2', 'Similarity'])
    return similar_titles_df

# Find and display similar titles
similar_titles_df = display_similar_titles(books)

# Print the DataFrame with similar titles
similar_titles_df
```

```python
def create_grouped_similar_books_dict(similar_titles_df):
    # Dictionary to store similar books with one key
    similar_books_dict = {}

    # Set to track books that have already been added to the dictionary
    included_books = set()

    # Iterate through the similar_titles_df
    for _, row in similar_titles_df.iterrows():
        id1 = row['ID1']
        id2 = row['ID2']

        # If neither book has been included, create a new entry with id1 as the key
        if id1 not in included_books and id2 not in included_books:
            similar_books_dict[id1] = [id2]
            included_books.add(id1)
            included_books.add(id2)

        # If id1 is already in the dictionary, add id2 to its list
        elif id1 in similar_books_dict:
            similar_books_dict[id1].append(id2)
            included_books.add(id2)

        # If id1 is already included as a key but not in the dict (id2 was the key), skip id1
        elif id1 in included_books and id1 not in similar_books_dict:
            continue

        # If id2 is already in the dictionary, add id1 to its list
        elif id2 in similar_books_dict:
            similar_books_dict[id2].append(id1)
            included_books.add(id1)

        # If id2 is already included but not as a key (id1 was the key), skip id2
        elif id2 in included_books and id2 not in similar_books_dict:
            continue

    return similar_books_dict

# Create the dictionary for grouped similar books
grouped_similar_books_dict = create_grouped_similar_books_dict(similar_titles_df)

# Print the resulting dictionary
print(grouped_similar_books_dict)
```
